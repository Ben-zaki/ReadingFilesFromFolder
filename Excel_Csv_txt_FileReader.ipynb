#### appending files 

```python
#create an empty dataframe
dfempty = pd.DataFrame()
```

```python
for file in Csvfiles:
    df = pd.read_csv(file)
    dfglobal = pd.concat([dfempty,df])
    dfempty = dfglobal
```

```python
dfglobal.to_excel('CombinedSalesData.xlsx')
```
